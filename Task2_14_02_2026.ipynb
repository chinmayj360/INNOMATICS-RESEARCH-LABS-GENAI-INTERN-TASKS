PROGRAM 1 UNIQUE WORDS IN A SENTENCE 


```python
#take input from the user
sentence=input("Enter the sentence :")
#store the individual words in the form of list usinf split function 
words=sentence.split()
#push the entire list into a set to eliminate the duplicates
unique_words=set(words)
#the number of unique words is equal to the count of elements in the set as there is no duplicate entry in the sets
print("no of unique words",len(unique_words))
#printing the all the unque words
print("The list of uniques is as follows",unique_words)
```

PROGRAM 2 HIGHEST SALARY FROM EMPLOYEE DATA

```python
#emplty dictionary to store employee names and repective salaries
employees={}
n=int(input("Enter the number of employees!"))
#taking input for the salary entries
for i in range(n):
    name=input(f"Enter Employee {i+1} name?")
    salary=int(input(f"Enter salary of Employee {i+1} ?"))
    employees[name]=salary
highest_name=""
highest_salary=0
#comparing with prefined value 0 initially and with cuurentn higesh value until reaching the end of the of the dictionary 
for name,salary in employees.items():
    if salary>highest_salary:
        highest_salary=salary
        highest_name=name
print(F"HIGHEST SALARY IS ₹{employees[highest_name]} received by {highest_name}")
```

PROGRAM 3 FIND MAXIMUM AND MINIMUM VALUES

```python
#take user input in the form of string sepeated nny spaces
enter_list=input("enter the number seperated by sapace(single)")
#store the numerical values seperatly in a list of number by converting them to integer
final_list=list(map(int,enter_list.split()))
#for camparism keep the initial values of the max and min value as the first element of the list 
max_value=final_list[0]
min_value=final_list[0]
#traverse in tge list and update the max and min value when encountered a greater value and smaller value respectively
for num in final_list:
    if num>max_value:
        max_value=num
    elif num<min_value:
        min_value=num
#printing the solution 
print(f"the minimum value is {min_value}")
print(f"the maximum value is {max_value}")
print(final_list)
```

PROGRAM 4 COUNT PRODUCTS ABOVE A PRICE THRESHOLD

```python
#take user input in the form of string seperated by spaces
price=input("enter the prices seperted by spaces")
#store the values in the form of numbers in a list 
price_list=list(map(int,price.split()))
#counter variable to store number of item with cost greater than 1000
count=0
#iterate in the list and increament the count by one went encountered by value greater than 1000
for num in price_list:
    if num>1000:
        count+=1
print(f"the number of item with price greather than ₹1000 is {count}")
```

PROGRAM 5 FINDING ATTENDANCE PERCENTAGE

```python
#take user input in form of string seperated by space 
attendance_list=input("Enter attendance seperated by space(P for present and A for absent)")
#extract each char in a list 
final_list=attendance_list.split()
#counter to track the number of days present and absent
present=0
absent=0
#traverse in the list mark preseent if p is encountered , absent if a is encountered else throw error
for day in final_list:
    if day.upper()=='P':
        present+=1
    elif day.upper()=='A':
        absent+=1
    else:
        print("Invalid data..enter the correct data, may lead to incorrect results")
# print the solution 
percentage=round(present/(len(final_list)) *100 ,2 )     
print(f"The final attendance percentage is {percentage}")
```

PROGRAM 7 CHARACTER COUNT IN A STRING 

```python
#take user input 
text=input("enter the string to check characte frequency !")
#declare a python dictionary to keep a track of the char frequency 
char_list={}
#traverse in the string and increament the frequency of the encounterd charecter in the dicrionary 
for c in text:
    if c in char_list:
        char_list[c]+=1
    else:
        char_list[c]=1
#printing the char and its correcponding frequency 
print("the character and its frequency is as follows")
print(char_list)
```

PROGRAM 10 CALCULATE AVERAGE MARKS

```python
#taking the user input in the form of string
marks=input("Enter the marks of the students separated by space")
#extract the marks and store in the numerical list 
marks_list=list(map(int,marks.split()))
#counter to store score 
total=0

for num in marks_list:
    total+=num
avg=round(total/len(marks_list),2)
#Printing the solution 
print(f"The avgerage is {avg} considering the {len(marks_list)} entries")
```
