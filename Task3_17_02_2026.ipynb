Probelem Statement 1-Employee Performance Bonus Eligibility

```python
#taking user input
n=int(input("Enter the number of employees:"))
print("Enter the Employee details:\n")
Employee={}
high_val=0
#main logic, while taking input comparing it with max value if greater update the max value 
for i in range(0,n):
    name=input(f"Enter the name of Employee{i+1}")
    val=int(input(f"Enter the score of Employee{i+1}"))
    if val>high_val:
        high_val=val
    Employee[name]=val
final_list=list()
#storing the names of the employees with maximum salary
for name,val in Employee.items():
    if val==high_val:
        final_list.append(name)
print(f"The employees eligible for Bonus are {', '.join(final_list)} with score{high_val}")
    
    
    
```

Problem 2: Search Query Keyword Analysis

```python
import string
#taking user input
s=input("Enter the string")
#removing the punctuations from the string
s1=s.translate(str.maketrans("","",string.punctuation))
#storing the words in the list 
l=list(map(str.lower,s1.split()))
final_dic={}
#counting the frequency of the words using the dictionary
for word in l:
    if word not in final_dic:
        final_dic[word]=1
    else:
        final_dic[word]+=1
#printing the solution 
print("The frequency count is as follows:")
for word,count in final_dic.items():
    print(f"{word}: {count}")
    
```

PROBLEM 3 SENSOR DATA VALIDATION 

```python
#taking the sensor data in form of string seperate by spaces
sensor_data=input("Enter the sensor data seperated by space")
#storing the seperate sensor data in the form of integer in the list
sensor_list=list(map(int,sensor_data.split()))
final_list=list()
#traversing in the list to print only the even values
for i in range(0,len(sensor_list)):
    if i%2==1:
        final_list.append(sensor_list[i])
print("The reading of the valid hours are as follows:\n")
for i in range(0,len(final_list)):
    print(f"Hour {(i+1)*2}:reading {final_list[i]}")
```

Problem 4: Email Domain Usage Analysis

```python
#prelisted email ids'
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
#emplty dictionary to keep a track of the frequency of tthe domains listed
domain_count={}
#domain=list()
#main logic,traversing in the list and extracting the domain and keeping the track of the domain count using dictionary
for email in emails:
    #extracting the domain from the individual email id s'
    domain=email.split("@")[1]
    if domain not in domain_count:
        domain_count[domain]=1
    else:
        domain_count[domain]+=1
total_len=len(emails)
#printing the solution 
for mail,count in domain_count.items():
    print(f"domain:{mail}-> count:{count}")
    percentage=(count/total_len)*100
    print(f"{percentage}%")
    
```

```python
#taking the input from the users in the form of string 
user_id=input(f"Enter the user ids' seperated by spaces")
#storing the individual user names in the form of elements in the list 
user_list=list(user_id.split())
#dictionary to store the frequency of the  usernmaes
user_count={}
#main logic: considering each of the user name in the list and updating their frequency 
for name in user_list:
    if name not in user_count:
        user_count[name]=1
    else:
        user_count[name]+=1
#printing the output in user understandable way
print("The the user names and its correcsponding frequency is as follows")
for name,count in user_count.items():
    print(f"user name:{name}->count;{count}")
```
