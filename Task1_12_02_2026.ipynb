PROGRAM 2 PASS-FAIL ANALYSER

```python
# The list marks of the students given in the question 
marks = [45, 78, 90, 33, 60]
#declared two variables to store the count of students who have passed or failed
pass_count=0;
fail_count=0;
print("Analysing the marks list")
#main login : traversing in the marks list and categorising the students to pass or fail based on the give condition
#alongside keeping the count of pass and fail students
for i in range(0,len(marks)):
    if marks[i]>=50:
        print(f"Student {i+1} have passed with {marks[i]-50} over passsing marks")
        pass_count+=1;
    else:
        print(f"The student {i+1} have failed in the exam by {50-marks[i]} marks")
        fail_count+=1;
total_count=len(marks)
print("\n")
print(f"The total number of students is {total_count}")
print(f"Out of which {pass_count} students passed and {fail_count} students failed")
print("\n")
print(f"TOTAL STUDENTS = {total_count}")
print(f"PASS ={pass_count}")
print(f"FAIL ={fail_count}")
```

PROGRAM 3 SIMPLE DATA CLEANER

```python
#List of names given in the question 
names = [" Alice ", "bob", " CHARLIE "]
#decalring a new empty list to store the names after filtering considerijng the conditions given in the question 
filtered_names=[];
for name in names:
    new_name=name.strip().lower()
    #srip function removes the extra space and and lower function converts to lower case and the value is stored in a temporay varible
    #append the cleaned name to the new list 
    filtered_names.append(new_name)
print("The names after filtering are as follows")
for name in filtered_names:
    print(name)
```
