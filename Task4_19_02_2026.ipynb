PROBLEM 1-> Mobil Recharge Validation System

```python
def check_amount(amount):
    #valid list of recharge packs
    valid_list=[199,299,399,599]
    #checking whether greater than minimum val
    if amount<50:
        print("Alert,the amount is below the minimum fare ₹50.")
        return False
    elif amount not in valid_list:
        print("Warning, no such plan exist, ivalid plan ")
        print(f"The existing plans are {valid_list}")
        return False
    return True
while True:
    user_amount=int(input("Enter the amount for recharge!."))
    if check_amount(user_amount):
        print("Recharge sucessfull..")
        print(f"Enjoy the benefits of the plan ₹{user_amount}")
        break
    else:
        print("Please try again...")
```

TASK 2->Inventory reorder alert system 

```python
n=int(input("Enter the number of product"))
inventory={}
for i in range(0,n):
    product=input(f"Enter the product {i+1}")
    val=int(input(f"Enter the stock of product {i+1}"))
    inventory[product]=val
print("Analysing the data..")
#checking stock quantity for all the items in the inventory
for p,v in inventory.items():
    if v<15:
        print(f"{p}:reorder alert! ->(quantity={v})")
    else:
        print(f"{p}:stock OK ->(quantity={v})")

    
```

TASK 3-> Student Result Processing System

```python
def check_result(mark_list):
    #variable to store total marks
    total=0
    for marks in mark_list:
        total+=marks
    average=total/len(mark_list)
    if average>=50:
        result="Pass"
    else:
        result="Fail"
    return average,result
#predefined marks given in the question
marks = [60, 45, 70, 55, 40]

avg, final_result = check_result(marks)

print(f"Average Marks: {avg:.2f}")
print(f"Final Result: {final_result}")
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def cab_fare(distance, peak_hour):
    base_fare = 50 #base fare fixed and independent of the distance travelled
    unit_fare = 12 #fare per unit kilometer
    total_fare = base_fare + (distance * unit_fare)
    #check if  it is peak hour if yes, apply extra fare 
    if peak_hour:
        total_fare *= 1.25
        
    return total_fare


while True:
    d = int(input("Enter distance travelled (km): "))
    peak = input("Is it peak hour (yes/no)? ")

    cost = cab_fare(d, peak.lower() == "yes")
    print(f"The total cab fare is ₹{cost}")

    retry = input("Do you want to retry? (yes/no): ")
    if retry.lower() == "no":
        print("Thank you for using the Cab Fare Estimator!")
        break
```

Problem 5: Employee Attendance Eligibility Checker


```python
def attendance_check(attendance):
    #conerting the entry from string to list 
    attendance_list=attendance.split()
    #data cleaning, terminating program if invaild entry in encountered
    for day in attendance_list:
        if day.lower() not in ['p','a']:
            print("Invalid Data")
            return
    total_days=len(attendance_list)
    present_days=0
    for day in attendance_list:
        if day.lower()=='p':
            present_days+=1
    print("Analysing the attendance..")
    print(f"The total number of days present {present_days}")
    print(f"The total number of working days is {total_days}")
    #applying logic to calculate the final percentage
    print(f"The attendance percentage is {round(((present_days/total_days)*100),2)}")
    return "Eligible" if (present_days/total_days)*100 >= 75 else "Not Eligible"
attendance=input("Enter the attendance sepeated by spaces!")
result=attendance_check(attendance)
print(f"Employee is {result}")
```

Problem 6 Password strength checker

```python
def check_password(password):
    #checking the length of the password
    if len(password)<8:
        return f"Weak Password must contain at least 8 characters"
    #boolean variables to check the availiability of atleast one digit and at least one speacil character("@#$")
    chk_digit=False
    chk_special=False
    #traverse in the string to to check it satisfies the condition 
    for p in password:
        if p.isdigit():
            chk_digit=True
        if p in "@#$":
            chk_special=True
    if chk_digit==False:
        return f"WeaK password!, the password must contain atleast one digit"
    if not chk_special:
        return f"Weak password !,the password must contain atlest one special character"
    return "strong password"
passkey=input("Enter the password!")
result=check_password(passkey)
print(result)
```
