PROGRAM 1  Smart Light Controller

```python
class Smart_Light:
    def __init__(self,name):
        self.name=name
        self.status="OFF"
    def turn_on(self):
        self.status="ON"
    def turn_off(self):
        self.status="OFF"
    def show_status(self):
        print(f"{self.name} is {self.status}")
smart1=Smart_Light("Light1")
smart1.turn_on()
smart1.show_status()
```

Problem 4: Product Price Tag Generator

```python
class Product_Tag:
    def __init__(self,name,cost):
        self.name=name
        self.cost=cost
    def show_tag(self):
        print(f"Prodcut:{self.name}")
        print(f"Cost:₹{self.cost}")
products=[]
p1=Product_Tag("Earphones",4500)
p2=Product_Tag("Table fan",8500)
p3=Product_Tag("Brake pads",300)
products.append(p1)
products.append(p2)
products.append(p3)
for p in products:
    p.show_tag()
```

Problem 5: Movie Rating Display System

```python
class Movies_Rating:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
    def show_rating(self):
        print(f"Movie : {self.name}     rating:{self.rating}/5 stars")
movies=[]
n=int(input("Enter the number of movies.."))
for i in range(0,n):
    name=input(f"Enter the name of movie{i+1}")
    rating=float(input("Enter rating"))
    m=Movies_Rating(name,rating)
    movies.append(m)
for m in movies:
    m.show_rating()
    
```

Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self,name,address):
        self.name=name
        self.address=address
    def show_add(self):
        print(f"Name:{self.name}")
        print(f"Address:{self.address}")
n=int(input("Enter the number of clients.."))
clients=[]
for i in range(0,n):
    name=input(f"Enter the name of client {i+1}")
    address=input(f"Enter the address of client {i+1}")
    d=Delivery(name,address)
    clients.append(d)
for c in clients:
    c.show_add()
    
```
