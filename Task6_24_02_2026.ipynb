PROBLEM 1 Smart Parking Lot Management System

```python
def parking(peak_capacity,p_list):
    count=0#counter to store the number of vehicals in the lot currently
    max_count=-1
    for p in p_list:
        if p=="IN":
            count+=1
        elif p=="OUT" and count>0:
            count-=1
        max_count=max(count,max_count)
        if count==peak_capacity:
            print(f"Max capacity reached..Parking slot is full")
            break
    return count
peak=int(input("Enter the peak capacity of the parking lot:"))
p_list=["IN","OUT","IN","IN","OUT","IN"]
ans=parking(peak,p_list)
print(f"Current parking slot taken is {ans}")
```

 Problem Statement 2:Online Food Delivery Time Estimator


```python
def fare_cal(distance,weather,traffic):
    time=0
    w_time=0
    t_time=0
    #considering the average speed of travel as 20km/h
    #using the formula time=distance/speed
    time+=(distance/20)*60 #comsidering the time in minutes
    if weather=="CLEAR":
        pass
    if "RAINY"in weather:
        w_time+=(time*0.15) #buffer of 15 % extra time
    if "FOG" in weather:
        w_time+=(time*0.18) #buffer of 18 % extra time
    if "WINDY" in weather:
        w_time+=(time*0.19) #buffer of 19 % exttra time
    if "STORMY" in weather:
        w_time+=(time*0.20) #buffer of 20 % extra time
    if traffic=="LOW":
        t_time=-1*(time*0.10) #you can reduce time when the traffic is low
    elif traffic=="MEDIUM":
        t_time=(time*0.05) #additional time on moderate traffic
    elif traffic=="HIGH":
        t_time=(time*0.35) #additional time for high traffic
    time+=w_time+t_time
    time=time*1.1 # considering 10 % buffer
    hr=int(time//60)
    mini=int(round(time%60,0))
    return hr,mini
hr,mini=fare_cal(45,["RAINY","WINDY","STORMY"],"LOW")
print(f"The estimated time is  {hr} hours:{mini} mins")
        
```

 Problem Statement 3:Movie Theatre Seat Occupancy

```python
def seat_cal(s_list,capacity):
    num=0#counter to store the val containing the no of people in the event currently
    for s in s_list:
        if num+s>capacity:
            print("No sufficient seats for the transaction..try next show")
            continue
        num+=s
        if num==capacity:
            print("HOUSE FULL ,MAXIMUM CAPACITY REACHED ")
            break
    STATUS=""
    percentage=round((num/capacity)*100,2)
    if percentage>0 and percentage<50:
        STATUS="SEATS AVAILABLE"#updating status
    elif percentage>=50 and percentage<=75:
        STATUS="FAST FILLING"#updating status
    elif percentage>75 and percentage<100:
        STATUS="ALMOST FULL"#updating status
    else:
        STATUS="HOUSE FULL"
    return num,STATUS
tickets=input("Enter the the tickects pUrchaces request in a line seperated by space")
tickets_list=list(map(int,tickets.split()))
capacity=int(input("Enter the maximum capacity"))
num,STATUS=seat_cal(tickets_list,capacity) 
print(f"{num} seats out of {capacity} are occupied..")
print(STATUS)
```


 Problem Statement 4:Cloud Server Load Classification System

```python
def server_load(cpu_readings):
    # Calculate average CPU usage
    total = 0 #counter to store total cpu usage 
    for reading in cpu_readings:
        total += reading
    #calculating the average cpu reading in the given list
    avg_cpu = total / len(cpu_readings)
    print(f"Average CPU Load: {int(avg_cpu)}%")
    # Classification
    if avg_cpu < 50:
        status = "Normal"
    elif 50 <= avg_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"
    print(f"Server Status: {status}")
# Sample Input
cpu_readings = [45, 60, 70, 85, 90]
server_load(cpu_readings)
```

Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def resource_monitor(usage_dict):
    overused = []   
    # Define maximum recommended hours per day
    limit = {
        "Projector": 5,
        "AC": 8,
        "Lights": 6
    }   
    # Check each resource
    for resource in usage_dict:
        if resource in limit:
            if usage_dict[resource] > limit[resource]:
                overused.append(resource)    
    # Output
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

 Problem Statement 6: Online Event Registration Capacity Controller


```python
def event_mng(capacity,e_list):
    count=0 #conter to count the no of registered events
    w_count=0 #counter to count the no of waiting list events
    #counting the total no of evnts registered
    for e in e_list:
        count+=e
    if count>=capacity:
        w_count=count-capacity
        count=capacity
        STATUS="REGISTERATION CLOSED"
    else:
        STATUS="REGISTERATION OPEN,SLOTS AVAILABE"
    return count,w_count,STATUS
events=input("Enter the events booking separeted by spaces..")
events_list=list(map(int,events.split()))
capacity=int(input("Enter the maximum capacity of the event manager"))
count,w_count,STATUS=event_mng(capacity,events_list)
print(f"{count} events booked. {w_count} in the waiting list")
print(f"STATUS : {STATUS}")
        
```
