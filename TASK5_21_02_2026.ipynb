Problem 1: Social Media – Post Engagement Analyzer

```python
def likes_analyse(likes_list):
    #counter to store the total no of likes
    total=0
    for like in likes_list:
        total+=like
    #main logic to classify the post to viral or normal enagement category
    if total>=1000:
        print(f"Viral post..")
    else:
        print(f"Normal enagement..")
likes=input("Enter the likes seperated by space")
likes_list=list(map(int,likes.split()))
likes_analyse(likes_list)
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def rainfall_analyser(rain_list,reqlvl):
    total=0
    for rain in rain_list:
        total+=rain
    #main logic to calculate the average
    avg_rain=total/len(rain_list)
    print(f"Average rainfall is {avg_rain} units")
    #classification of the rainfall based on the average rainfall values
    if avg_rain>=reqlvl:
        print("STATUS : ADEQAUTE RAINFALL THIS SEASON")
    else:
        print("IN sufficient rainfall this season")
rain_data=input("Enter the rain fall in mm seperated by spaces")
rain_list=list(map(int,rain_data.split()))
req_level=int(input("Enter the required rain level for reference in mm"))
rainfall_analyser(rain_list,req_level)
```

TASK 4 Social Media – Duplicate Account Detection


```python
def check_duplicates(user_list):
    #entering the list vaues into sets to elimate the duplicate entry
    user_set=set(user_list)
    #main logic to check the existance of duplicates
    if len(user_list)!=len(user_set):
        print("Yes ,duplicate accounts found..")
    else:
        print("No duplicate account found, all the accounts in the data base are unique")
users=input("Enter the user list separated  by ,")
user_list = [user.strip() for user in users.split(",")]
check_duplicates(user_list)
```

Problem 6: Agriculture – Premium Crop Price Filter


```python
def crop_filter(crops):
    for name,price in crops.items():
        #main logic to classify based on crop prices
        if price>2000:
            print(f"CROP {name} is a primium crop with cost ₹{price} per 100kg")
        else:
            print(f"Crop {name} is a regular crop with ₹{price} per 100kg")
crop={"Potato":1788,"Tomato":1899,"Onion":1888,"Avacado":2899}
crop_filter(crop)
```

Problem 7: System Monitoring – Application Health 

```python
def check_application_health(error_count):
    
    print("Error Count:", error_count)
    #main logic to classify the system status based on the error counts
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Taking input
errors = int(input("Enter error count: "))
check_application_health(errors)
```

Problem 9: E-Learning – Student Attendance Eligibility 

```python
def attendance_cal(attendance_list):
    #total number of classes condicted 
    total=len(attendance_list)
    present=0
    #counting the no of classes present
    for day in attendance_list:
        if day.lower()=='p':
            present+=1
    #calculating the present percentage
    percentage=round((present/total)*100,2)
    #main logic to classify the student present or absent based on the attendance persentage
    if percentage>=75:
        print("Student is eligible to attend the exams")
    else:
        print("Student is not eligible to attend the exam")
    return total,percentage,present
att=input("Enter the attance entry seperated by ,")
att_list=[atti.strip() for atti in att.split(",")]
check=True
#terminating the program when there is a error encountered
for atti in att_list:
    if atti.lower() not in ['p','a']:
        print("Invalid entry")
        check=False
        break
if check:
    total,percentage,present=attendance_cal(att_list)
    print(f"TOTAL DAYS:{total}")
    print(f"NO OF DAYS PRESENT:{present}")
    print(f"ATTENDANCE PERCENTAGE:{percentage}")
```

Problem 10: Smart Electricity Bill Analyzer

```python
def bill(units):
    #returning error when the number of units is less than 0
    if units<0:
            status="ERROR"
            cost=0
            print("Exception handler :INVALID ENTRY")
            return cost,status
    # main logic to calculate the bill amount and classifycation of the bill status
    elif units<=100:
        cost=units*3
        status="Low usage"
        return cost,status
    elif units<=200:
        cost=(300+(units-100)*5)
        status="Low usage"
        if cost>500:
            status="Moderate usage"
        return cost,status
    else:
        cost=300+500+(units-200)*7
        status="Moderate usage"
        if cost>1500:
            status="High usage"
        return cost,status
units=int(input("Enter the no of units used"))
b_cost,b_status=bill(units)
print(f"BIll  COST:₹{b_cost}")
print(f"USAGE STATUS :{b_status}")
```
